```python
#TP4 : Les fractions "égyptiennes"
from fractions import *
def fraction_egyptiennes(a = 2, b = 17):
    q = 1
    while( a*q < b ):
        q = q + 1
#     print(q)
    res = Fraction(a, b)-Fraction(1,q)
#     print(res)
    if res.numerator == 1:
        print( Fraction(a, b)," = ", Fraction(1, q), " + ", res)
    else:
        print("Décomposition incomplétentes, il faut plus de deux fractions égyptiennes")

fraction_egyptiennes()
fraction_egyptiennes(2,5)
fraction_egyptiennes(3,4)
fraction_egyptiennes(3,11)
fraction_egyptiennes(4,11)
fraction_egyptiennes(5,9)
fraction_egyptiennes(8,9)
    
```
