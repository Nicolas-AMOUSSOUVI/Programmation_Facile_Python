```python
#TP 17 Variance et écartètype
from math import*

def moyenne(A):
    res = 0
    for i in range(len(A)):
        res = res + A[i]
    res = res/(len(A))
    return res

def variance(A):
    res = 0
    m = moyenne(A)
    for i  in range(len(A)):
        res = res + (A[i] - m)**2
    res = res/(len(A))
    return res

def ecartype(A):
    return sqrt(variance(A))

A = [10,5,7,15,18,19]
moyenne(A)
variance(A)
ecartype(A)
```

```python
#TP 20 Tours de Hanoï

def tour_dHanoi(n = 3, D = 1 , A = 2, I = 3 ):
    #n nombre de disque
    #D emplacement de départ
    #A emplacement d'arrivée
    #I emplacement intermédiaire
    
    if n == 1:
        print("de",D,"vers",A)
    else:
        tour_dHanoi(n-1,D,I,A)
        tour_dHanoi(1, D, A, I)
        tour_dHanoi(n-1, I, A, D)

tour_dHanoi()
tour_dHanoi(10)
```
